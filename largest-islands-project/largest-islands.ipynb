```python
# Get the top 10 largest islands
top_10_islands = (
    df.sort_values(by="area", ascending=False)
      .head(10)
)


top_10_islands
```

```python
#Largest Islands
# Largest island in each region
largest_by_region = (
    df.loc[df.groupby('region')['area'].idxmax()]
    .sort_values(by='region')
)

largest_by_region
```

```python
#Visualisation
# Sort by rank
ranked_islands = df.sort_values(by="rank")

plt.figure(figsize=(12,6))

plt.plot(
    ranked_islands['rank'],
    ranked_islands['area'],
    color = 'red',
    marker='o'
)

plt.title("Island Area by Rank")
plt.xlabel("Rank")
plt.ylabel("Area (km²)")

#plt.grid(True)
plt.show()
```

```python
# Islands belonging to multiple countries
multi_country_islands = df[
    df['countries'].str.contains(",", na=False)
]

multi_country_islands
```
